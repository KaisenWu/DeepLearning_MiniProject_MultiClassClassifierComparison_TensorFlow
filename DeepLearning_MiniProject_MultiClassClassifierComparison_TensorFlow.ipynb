## Data Loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Lab5/dataset4.csv')
```

```python
df.info()
```

## Split and scale the data.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('class', axis=1).values
y = df['class'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Build The Model

```python
layer_list1 = [2,5,10]
layer_list2 = [2,5,10]
layer_list = []
for layer1 in layer_list1:
  for layer2 in layer_list2:
    if layer1 >= layer2:
      layer_list.append([layer1, layer2])
layer_list
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

```python
models = []
for layer in layer_list:
  model = Sequential()
  model.add(Dense(layer[0],activation='relu'))
  model.add(Dense(layer[1],activation='relu'))
  model.add(Dense(3,activation='softmax'))
  model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
  models.append(model)
```

## Model Training

```python
for model in models:
  model.fit(x=X_train, y=y_train, epochs=200, validation_data=(X_test, y_test),callbacks=[early_stop])
```

## Model Evaluation

```python
for model in models:
  model.evaluate(X_test, y_test, verbose=1)
```
